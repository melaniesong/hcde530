**Question 1:** Which Asian cuisines (e.g., Chinese, Japanese, Thai, Vietnamese, etc.) are most and least represented among restaurants in the dataset?


```python
# Your analysis for Question 1
# Question 1:
# Which Asian cuisines are most and least represented?

import pandas as pd

# Load cleaned file (created earlier)
df = pd.read_csv("restaurants_cleaned.csv", low_memory=False)

# 1) Keep only rows with a category
work = df.copy()
work["Category"] = work["Category"].fillna("").astype(str).str.strip()
work = work[work["Category"] != ""]

# 2) Map raw category labels -> cuisine groups
# (edit/expand this mapping if you want to include more labels)
cuisine_map = {
    "Chinese": [
        "Chinese restaurant", "Chinese takeaway", "Chinese noodle restaurant",
        "Chinese bakery", "Chinese tea house", "Delivery Chinese restaurant"
    ],
    "Japanese": [
        "Japanese restaurant", "Authentic Japanese restaurant",
        "Japanese steakhouse", "Japanese curry restaurant",
        "Japanese regional restaurant", "Japanese delicatessen",
        "Kyoto style Japanese restaurant"
    ],
    "Thai": ["Thai restaurant"],
    "Vietnamese": ["Vietnamese restaurant"],
    "Korean": [
        "Korean restaurant", "Korean barbecue restaurant", "Korean rib restaurant"
    ],
    "Indian": [
        "Indian restaurant", "North Indian restaurant", "South Indian restaurant",
        "North Eastern Indian restaurant", "Modern Indian restaurant",
        "Indian Muslim restaurant", "Indian takeaway"
    ],
    "Filipino": ["Filipino restaurant"],
    "Malaysian": ["Malaysian restaurant"],
    "Indonesian": ["Indonesian restaurant"],
    "Taiwanese": ["Taiwanese restaurant"],
    # Optional broader buckets:
    "Pan-Asian / General Asian": [
        "Asian restaurant", "Asian fusion restaurant",
        "Pan-Asian restaurant", "South Asian restaurant", "Southeast Asian restaurant"
    ],
}

# Build reverse lookup (raw category -> cuisine group)
reverse_map = {}
for cuisine, labels in cuisine_map.items():
    for label in labels:
        reverse_map[label.lower()] = cuisine

# 3) Assign cuisine group
work["Cuisine_Group"] = work["Category"].str.lower().map(reverse_map)

# Keep only mapped Asian cuisines
asian = work[work["Cuisine_Group"].notna()].copy()

# 4) Count representation
counts = (
    asian["Cuisine_Group"]
    .value_counts()
    .rename_axis("Cuisine")
    .reset_index(name="Restaurant_Count")
)

# Add percent share within Asian subset
counts["Percent_of_Asian"] = (counts["Restaurant_Count"] / counts["Restaurant_Count"].sum() * 100).round(2)

# Most represented
print("Most represented cuisines:")
display(counts.head(5))

# Least represented (non-zero)
print("Least represented cuisines:")
display(counts.tail(5))

# Full table
display(counts)
```

**Interpretation:**  
Asian cuisine representation in this dataset is uneven. Chinese restaurants are the most represented category by a wide margin, followed by Thai and Japanese, while cuisines such as Indonesian, Malaysian, and Filipino appear much less frequently. This suggests that some Asian cuisines are much more visible on this platform than others, which could reflect differences in real-world restaurant prevalence, platform coverage, or category labeling practices. A key limitation is that cuisine labels are not perfectly standardized (for example, broad labels like “Asian restaurant” and mixed labels like “Asian fusion”), so these counts should be interpreted as approximate representation rather than exact cultural market share.


**Question 2:** How does the representation of Asian cuisines vary across different U.S. states (e.g., which states have higher or lower proportions of Asian cuisine restaurants)?


```python
# Your analysis for Question 2
# Question 2:
# How does representation of Asian cuisines vary across U.S. states?

import pandas as pd

# Load cleaned dataset
df = pd.read_csv("restaurants_cleaned.csv", low_memory=False)

# -----------------------------
# 1) Create an Asian flag
# -----------------------------
pattern = r"\b(?:asian|chinese|japanese|thai|vietnamese|korean|indian|filipino|malaysian|indonesian|taiwanese)\b"

work = df.copy()
work["Category"] = work["Category"].fillna("").astype(str).str.strip()
work["is_asian"] = work["Category"].str.lower().str.contains(pattern, regex=True)

# Exclude false positive explicitly
work.loc[work["Category"].str.lower() == "caucasian restaurant", "is_asian"] = False

# -----------------------------
# 2) Extract U.S. state from Address
# -----------------------------
# Example format: "... , Seattle, WA 98101, United States"
work["Address"] = work["Address"].fillna("").astype(str)
work["State"] = work["Address"].str.extract(
    r",\s*([A-Z]{2})\s+\d{5}(?:-\d{4})?,\s*United States\s*$"
)[0]

# Keep only rows where state was extracted
state_df = work[work["State"].notna()].copy()

# -----------------------------
# 3) State-level representation
# -----------------------------
state_summary = (
    state_df.groupby("State", as_index=False)
    .agg(
        total_restaurants=("State", "size"),
        asian_restaurants=("is_asian", "sum")
    )
)

state_summary["asian_share_pct"] = (
    state_summary["asian_restaurants"] / state_summary["total_restaurants"] * 100
).round(2)

# Optional: require minimum sample size for fair comparison
state_summary_filtered = state_summary[state_summary["total_restaurants"] >= 500].copy()

# Highest and lowest shares
highest = state_summary_filtered.sort_values("asian_share_pct", ascending=False).head(10)
lowest = state_summary_filtered.sort_values("asian_share_pct", ascending=True).head(10)

print("Top 10 states by Asian restaurant share (min 500 total restaurants):")
display(highest)

print("Bottom 10 states by Asian restaurant share (min 500 total restaurants):")
display(lowest)
```

**Interpretation:**  
Representation of Asian cuisines varies substantially across states. Some states have a much higher proportion of Asian cuisine restaurants, while others have relatively low shares, indicating uneven geographic visibility and concentration. This suggests that local demographics, migration patterns, and regional food markets likely influence cuisine representation. A limitation is that state extraction depends on address formatting and category labels, so results are best interpreted as comparative patterns rather than exact market estimates.

**Question 3:** Do average ratings differ across Asian cuisine types, and are certain cuisines consistently rated higher or lower than others?

```python
# Your analysis for Question 3
# Question 3:
# Do average ratings differ across Asian cuisine types?

import pandas as pd

# Load cleaned dataset
df = pd.read_csv("restaurants_cleaned.csv", low_memory=False)

# Keep needed columns and clean
work = df[["Category", "Rating"]].copy()
work["Category"] = work["Category"].fillna("").astype(str).str.strip()
work["rating_num"] = pd.to_numeric(work["Rating"], errors="coerce")

# Keep only valid ratings
work = work[work["rating_num"].notna()]

# Map raw category labels -> cuisine groups
cuisine_map = {
    "Chinese": [
        "Chinese restaurant", "Chinese takeaway", "Chinese noodle restaurant",
        "Chinese bakery", "Chinese tea house", "Delivery Chinese restaurant"
    ],
    "Japanese": [
        "Japanese restaurant", "Authentic Japanese restaurant",
        "Japanese steakhouse", "Japanese curry restaurant",
        "Japanese regional restaurant", "Japanese delicatessen",
        "Kyoto style Japanese restaurant"
    ],
    "Thai": ["Thai restaurant"],
    "Vietnamese": ["Vietnamese restaurant"],
    "Korean": [
        "Korean restaurant", "Korean barbecue restaurant", "Korean rib restaurant"
    ],
    "Indian": [
        "Indian restaurant", "North Indian restaurant", "South Indian restaurant",
        "North Eastern Indian restaurant", "Modern Indian restaurant",
        "Indian Muslim restaurant", "Indian takeaway"
    ],
    "Filipino": ["Filipino restaurant"],
    "Malaysian": ["Malaysian restaurant"],
    "Indonesian": ["Indonesian restaurant"],
    "Taiwanese": ["Taiwanese restaurant"],
    "Pan-Asian / General Asian": [
        "Asian restaurant", "Asian fusion restaurant",
        "Pan-Asian restaurant", "South Asian restaurant", "Southeast Asian restaurant"
    ],
}

# Reverse mapping
reverse_map = {}
for cuisine, labels in cuisine_map.items():
    for label in labels:
        reverse_map[label.lower()] = cuisine

work["Cuisine_Group"] = work["Category"].str.lower().map(reverse_map)

# Keep mapped cuisines only
asian_ratings = work[work["Cuisine_Group"].notna()].copy()

# Aggregate stats by cuisine
q3_summary = (
    asian_ratings.groupby("Cuisine_Group", as_index=False)
    .agg(
        n_restaurants=("rating_num", "size"),
        avg_rating=("rating_num", "mean"),
        median_rating=("rating_num", "median"),
        std_rating=("rating_num", "std")
    )
)

# Keep cuisines with enough observations for stability
q3_summary = q3_summary[q3_summary["n_restaurants"] >= 50].copy()

# Sort by average rating
q3_summary = q3_summary.sort_values("avg_rating", ascending=False)
q3_summary["avg_rating"] = q3_summary["avg_rating"].round(3)
q3_summary["median_rating"] = q3_summary["median_rating"].round(3)
q3_summary["std_rating"] = q3_summary["std_rating"].round(3)

display(q3_summary)

print("Highest average-rated cuisines:")
display(q3_summary.head(5))

print("Lowest average-rated cuisines:")
display(q3_summary.tail(5))
```

**Interpretation:**  
Average ratings do differ across Asian cuisine types, but the differences are relatively modest on a 1–5 scale. Some cuisine groups consistently appear near the top of average ratings, while others are lower, suggesting variation in customer satisfaction (or reviewing behavior) across cuisine categories. To avoid over-interpreting small groups, I applied a minimum sample-size threshold (n >= 50) before comparing means. A limitation is that ratings are platform-generated and may be influenced by location, chain effects, and review volume, so these results show comparative patterns rather than causal quality differences.

```python
#cholopleth of asian restaurants by state
import pandas as pd
import plotly.express as px

# --- Load data ---
df = pd.read_csv("restaurants_cleaned.csv", low_memory=False)

# --- Asian flag (word-boundary; excludes Caucasian false positive) ---
pattern = r"\b(?:asian|chinese|japanese|thai|vietnamese|korean|indian|filipino|malaysian|indonesian|taiwanese)\b"
work = df.copy()
work["Category"] = work["Category"].fillna("").astype(str).str.strip()
work["is_asian"] = work["Category"].str.lower().str.contains(pattern, regex=True)
work.loc[work["Category"].str.lower() == "caucasian restaurant", "is_asian"] = False

# --- US state from address: ", ST 12345, United States" ---
work["Address"] = work["Address"].fillna("").astype(str)
work["State"] = work["Address"].str.extract(
    r",\s*([A-Z]{2})\s+\d{5}(?:-\d{4})?,\s*United States\s*$"
)[0]

us = work[work["State"].notna()].copy()

# --- Aggregate: Asian restaurant count per state ---
choropleth = (
    us.groupby("State", as_index=False)
      .agg(
          total_restaurants=("State", "size"),
          asian_restaurants=("is_asian", "sum"),
      )
)

choropleth["asian_share_pct"] = (
    choropleth["asian_restaurants"] / choropleth["total_restaurants"] * 100
).round(2)

# --- Choropleth: COUNT of Asian restaurants ---
fig = px.choropleth(
    choropleth,
    locations="State",
    locationmode="USA-states",
    color="asian_restaurants",
    scope="usa",
    title="California has the most number of Asian cuisine restaurants purely by count",
    labels={"asian_restaurants": "Asian restaurants"},
    color_continuous_scale="Blues",
)
fig.update_layout(
    margin=dict(l=0, r=0, t=50, b=0),
    geo=dict(scope="usa", projection_scale=1.05),
)
fig.show()
fig.write_image("asian_cuisine_restaurants_by_state.png")
```

```python
#cholopleth by percent
fig2 = px.choropleth(
    choropleth,
    locations="State",
    locationmode="USA-states",
    color="asian_share_pct",
    scope="usa",
    title="Asian cuisine’s share of each state’s listings tells a different story than raw counts alone",
    labels={"asian_share_pct": "Asian share (%)"},
    color_continuous_scale="Reds",
)
fig2.update_layout(margin=dict(l=0, r=0, t=50, b=0), geo=dict(scope="usa"))
fig2.show()
fig2.write_image("share_of_restaurants_that_are_asian_cuisine.png")
```

**Chart rationale:**  
I used two choropleth maps side by side because they answer different but related questions about the same geography. The first map plots the number of Asian-cuisine restaurants per state; it highlights where the largest totals appear (often states with bigger populations and more listings overall). The second map plots share (% of restaurants in each state flagged as Asian cuisine); it adjusts for scale and makes it easier to compare whether Asian cuisine listings are unusually common or uncommon in smaller states versus larger ones.

Together, these views reduce misinterpretation: a state can rank high on counts but only moderate on share, meaning “many listings” partly reflects marketplace size—not only cultural food presence. Readers should interpret both maps relative to limitations in the dataset (category labels as proxies for cuisine and address-based state parsing dropping some rows). The takeaway should combine both signals: geographic concentration and proportional representation within each state’s slice of the data.
